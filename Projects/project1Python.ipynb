```python
# Import needed libraries 
from prettytable import PrettyTable
from random import randint 
```

## Products of the Store

```python
def display_products(products):
    table = PrettyTable()
    table.field_names = ["Name", "Price", "Quantity"]
    for product in products:
        table.add_row([product["Name"], product["Price"], product["Quantity"]])
    print(table)
```

## User Log in

```python
def user_login():
    OTP = randint(10000, 100000)
    user = False
    while not user:
        username = input("Enter username: ")
        password = input("Enter password At least 8 characters: ")
        if  username and len(password) >= 8:
            user = True
            print(f"Your verification is {OTP}")
            OTP_user =  input("Enter the verification code")
            OTP_user = int(OTP_user)
            if OTP_user != OTP:
                print("Not Correct!")
                OTP_user = input("Enter again: ")
                OTP_user = int(OTP_user)

            else:
                print("verification code is correct!")
        else:
            print("User Name or Password is not Correct!")
            username = input("Enter User name: ")
            password = input("Enter Password at least 8 characters: ")
    else:
        print("User Successfully logged in!")
    
    return username, password
```

```python
# def order(product):
#     exists = False
#     indexOfPr = 0
#     while not exists:
#         for i in range(len(products)):
#             if product.capitalize() == products[i]["Name"]:
#                 exists = True
#                 indexOfPr = i
#                 break
#         else:
#             print("there is no product with this name!")
#             product = input("Enter the Product agian: ")
#     return product, indexOfPr


# def quantityOfOrder(quantity, i, availableQ):
#     availablilty = False
#     while not availablilty:
#         if quantity <= availableQ:
#             availablilty = True
#             dis = quantity // 250
#             print(f"number of disc {dis}")
#             discount = 25 if dis == 0 else (5 * dis)
#             print(f"Final discount is {discount}%")
#             price = products[i]["Price"]
#             finalPrice =  price * (1 - (discount / 100))
#             print(f"original price is {price} \n final Price is {finalPrice} USD")
#             break
#         else:
#             print(f"Not available pls enter less than {availableQ}")
#             quantity = input(f"Enter the Quantity You need to order again: ")
#             quantity = int(quantity)
    
#     return finalPrice, quantity

def calculate_discount(quantity):
    discount_percentage = min(5 * (quantity // 250), 25)
    return discount_percentage

def update_product_list(products, product_name, purchase_quantity):
    for product in products:
        if product["Name"].lower() == product_name.lower():
            product["Quantity"] -= purchase_quantity

def purchase_product(products):
    total_price = 0

    while True:
        product_name = input("Enter product name: ")
        product_exists = next((product for product in products if product["Name"].lower() == product_name.lower()), None)
        if not product_exists:
            print("Product not found. Please try again.")
            continue

        quantity = int(input("Enter required quantity: "))
        if quantity > product_exists["Quantity"]:
            print("Quantity exceeds available stock. Please try again.")
            continue

        discount_percentage = calculate_discount(quantity)
        price_after_discount = product_exists["Price"] * quantity * (1 - discount_percentage / 100)
        print(f"Price after {discount_percentage}% discount: {price_after_discount:.2f} USD")

        update_product_list(products, product_name, quantity)
        total_price += price_after_discount
        print("total: ",total_price)

        another_item = input("Do you want to add another item? (yes/no): ").lower()
        if another_item == "no":
            break

    return total_price

def checkout(total_price):
    delivery_choice = input("Do you want delivery(200 USD) or pick-up(50 USD)? (dl/pu): ").lower()
    delivery_charge = 200 if delivery_choice == "dl" else 50
    total_price += delivery_charge

    currency_choice = input("Choose payment currency (USD, EUR, EGP): ").upper()
    conversion_rates = {"EUR": 0.91, "EGP": 48.83}
    if currency_choice in conversion_rates:
        total_price *= conversion_rates[currency_choice]
        print(f"Total price: {total_price:.2f} {currency_choice}")
    else:
        print(f"Total price: {total_price:.2f} USD")

    print("Your order is on the way.")
```

## stationary products of the store

```python
def stationary_store():
    stationary_products = [
        {"Name": "Notebook", "Price": 50.0, "Quantity": 500},
        {"Name": "Pen", "Price": 10.0, "Quantity": 1000},
        {"Name": "Pencil", "Price": 5.0, "Quantity": 1500},
        {"Name": "Eraser", "Price": 3.0, "Quantity": 800},
        {"Name": "Ruler", "Price": 20.0, "Quantity": 300}
    ]
    
    display_products(stationary_products)
    
    stationary_total_price = 0
    while True:
        product_name = input("Enter stationary product name: ")
        product_exists = next((product for product in stationary_products if product["Name"].lower() == product_name.lower()), None)
        if not product_exists:
            print("Product not found. Please try again.")
            continue

        quantity = int(input("Enter required quantity: "))
        if quantity > product_exists["Quantity"]:
            print("Quantity exceeds available stock. Please try again.")
            continue

        loyalty_discount = 2 * (quantity // 50)
        price_after_discount = product_exists["Price"] * quantity * (1 - loyalty_discount / 100)
        print(f"Price after {loyalty_discount}% loyalty discount: {price_after_discount:.2f} USD")

        product_exists["Quantity"] -= quantity
        stationary_total_price += price_after_discount

        another_item = input("Do you want to add another stationary item? (yes/no): ").lower()
        if another_item == "no":
            break

    return stationary_total_price
```

## Run The program

```python
def main():
    products = [
        {"Name": "Water", "Price": 80.0, "Quantity": 1200},
        {"Name": "Soda", "Price": 130.0, "Quantity": 1200},
        {"Name": "Chips", "Price": 75.0, "Quantity": 1200},
        {"Name": "Bread", "Price": 45.0, "Quantity": 1200},
        {"Name": "Eggs", "Price": 65.0, "Quantity": 1200}
    ]
    
    # Part 1 - Display product list
    display_products(products)

    # Part 2 - Login and Verification
    user_login()
    
    # Part 3 - Purchase process
    total_price = purchase_product(products)

    visit_stationary = input("Do you want to visit our new stationary store? (yes/no): ").lower()
    if visit_stationary == "yes":
        stationary_total_price = stationary_store()
        total_price += stationary_total_price

    # Checkout process
    checkout(total_price)

if __name__ == "__main__":
    main()
```
